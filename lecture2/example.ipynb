```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./아잉.csv")
data
```

```python
filtered = data[data['강의명'].str.contains('중급')] # 중급이 들어있는 강의만 가져옴
filtered
```

```python
filtered = filtered[~filtered['학기'].str.contains('여름학기')]
filtered = filtered[~filtered['학기'].str.contains('겨울학기')]
filtered
```

```python
filtered = filtered.drop(columns = ['학과', '이수학점', '학과', '학수번호', '강의시간', '강의명'])
filtered
```

```python
processed = filtered.replace('-', np.NaN) # 문자 대신 숫자로 바꿔줌
```

```python
processed['학습량'] = processed['학습량'].astype(float)
processed['난이도'] = processed['난이도'].astype(float)
processed['학점'] = processed['학점'].astype(float)
processed['성취감'] = processed['성취감'].astype(float)
processed['강의만족도'] = processed['강의만족도'].astype(float)
```

```python
processed.info()
```

```python
num_col = processed.loc[:, ['교수명', '학습량', '난이도', '학점', '성취감', '강의만족도']] #학기는 object라서 그래프에서 제외 (숫자가 아니므로)
```

```python
from pandas.plotting import parallel_coordinates

plt.rcParams['font.family'] = 'Malgun Gothic'

plt.figure(figsize = (20, 10))
parallel_coordinates(num_col, '교수명', alpha=0.5)
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize = (20, 6))
plt.xticks(rotation=270)
sns.barplot(data = num_col, x = '교수명', y = '학습량')
```

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.figure(figsize = (20, 6))
plt.xticks(rotation=270)
sns.barplot(data = num_col, x = '교수명', y = '난이도')
```

```python
ax = plt.rcParams['font.family'] = 'Malgun Gothic'
ax = plt.figure(figsize = (20, 6))
ax = plt.xticks(rotation=270)
ax = sns.barplot(data = num_col, x = '교수명', y = '학점', color= 'red', alpha=0.3)
ax = sns.barplot(data = num_col, x = '교수명', y = '학습량', color= 'green', alpha=0.3)
ax = sns.barplot(data = num_col, x = '교수명', y = '난이도', color= 'blue', alpha=0.3)
ax = sns.barplot(data = num_col, x = '교수명', y = '성취감', color= 'orange', alpha=0.3)
ax = sns.barplot(data = num_col, x = '교수명', y = '강의만족도', color= 'yellow', alpha=0.1)
plt.show()
```

```python
num_columns = num_col.columns
plt.figure(figsize = (20, 30))
for i in range(5):
    ax = plt.subplot(5, 1, i+1)
    ax = plt.xticks(rotation=270)
    ax = sns.boxplot(data = num_col, x ="교수명", y = num_columns[i+1])
```
