```python
import requests
from bs4 import BeautifulSoup as bs
# 고려대학교 사이트로부터 장학금 페이지에 request를 보내 사용함
url = 'https://ee.korea.ac.kr/community/undernotice.html?schType=subject&schText=%EC%9E%A5%ED%95%99%EA%B8%88'
response = requests.get(url)
response.encoding = 'utf-8'
print(response.text)
```

```python
html = response.text
soup = bs(html, 'html.parser')
```

```python
contents = soup.select('div.cont')# 홈페이지에서 div.cont에 해당하는 객체를 가져옴 홈페이지에서 f12을 통해 확인 가능 (태그이름. 클래스이름 )의 형식
len(contents) #contents의 개수를 알려줌
```

```python
links = [] #for 문을 통해 각각의 콘텐츠 정보에 접근 
for content in contents: #contents 안의 하나 content에 대하여: 
    date = content.select_one('span.date').text.split('.') #태그이름이 span 클래스 이름이 date인 정보를 수집해서 text를 .으로 분리해서 나누고 2023인 것만 저장
    if int(date[0])==2023: #select.one과 select의 차이 list로 저장 안하느냐 하느냐의 차이
        link = content.select_one('a')['href'] #a 태그의 href만 가져와라 
        print(link)
        links.append(link)
```

```python
links = links[1:]
links
```

```python
for i in links:
    url = 'https://ee.korea.ac.kr/community/'+i  #현재 갖고 있는 url만으로는 사이트에 접근할 수 없다 그래서 앞에 사이트 이름을 추가로 붙여서 사이트로 이동할 수 있게끔 함
    response = requests.get(url)
    response.encoding = 'utf-8'
    html = response.text
    soup = bs(html, 'html.parser')
    title = soup.select_one('p.tit').get_text() #title 정보와 text 정보를 찾아와서 각각의 title과 text에 저장한다
    print(title)
    text = soup.select_one('div.contents_info').get_text()
    print(text)
```

```python
f = open("./result.txt", "a", encoding = 'utf-8') #읽어온 것들을 위에 반복문과 달리 파일에 저장 (result.txt)
for i in links:
    url = 'https://ee.korea.ac.kr/community/'+i
    response = requests.get(url)
    response.encoding = 'utf-8'
    html = response.text
    soup = bs(html, 'html.parser')
    title = soup.select_one('p.tit').get_text()
    print(title)
    f.write(title+"\n")
    text = soup.select_one('div.contents_info').get_text()
    print(text)
    f.write(text)
    f.write("\n---------------\n")
f.close()
```
